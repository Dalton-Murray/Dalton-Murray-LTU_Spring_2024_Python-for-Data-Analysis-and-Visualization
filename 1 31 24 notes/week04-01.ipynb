```python
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Create a figure and an axes.
fig, ax = plt.subplots()

# Plot x versus y.
ax.plot(x, y)

# Set a title and labels
ax.set_title('Simple Plot')
ax.set_xlabel('x values')
ax.set_ylabel('y values')

# Show the plot
plt.show()
```

```python
import numpy as np
fig, ax = plt.subplots()

# Using a Python list
x_list = [1, 2, 3, 4]
y_list = [1, 4, 9, 16]
ax.plot(x_list, y_list, label='Data from list')

# Using a NumPy array
x_array = np.array([1, 2, 3, 4])
y_array = np.array([1, 4, 9, 16])
ax.plot(x_array, y_array, label='Data from array', linestyle='--')

# Adding labels and title
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_title('Plotting with Lists and Numpy Arrays')
ax.legend()

# Display the plot
plt.show()
```

```python

# Customize the line style and color
plt.plot(x, y, color='red', linestyle='--', linewidth=2,label='Data')

# Add labels and title
plt.xlabel('X Axis Label')
plt.ylabel('Y Axis Label')
plt.title('Customized Line Plot')

# Show the plot with grid
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Create a figure and a 2x1 grid of subplots
fig, (ax1, ax2) = plt.subplots(2, 1)

# Plot on the first subplot
ax1.plot([0, 1, 2], [10, 20, 30])
ax1.set_title('First Subplot')

# Plot on the second subplot
ax2.plot([0, 1, 2], [30, 20, 10])
ax2.set_title('Second Subplot')

# Show the figure with subplots
plt.tight_layout()  # Adjust the layout so the titles do not overlap
plt.show()
```

```python
# Create 2x2 layout of subplots
fig, axs = plt.subplots(2, 2)

# Plot different data on each subplot
axs[0, 0].plot([0, 1, 2], [10, 20, 30])
axs[0, 1].plot([0, 1, 2], [30, 20, 10])
axs[1, 0].bar([0, 1, 2], [20, 10, 30])
axs[1, 1].scatter([0, 1, 2], [30, 10, 20])

# Customize the layout
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.4, hspace=0.6)

# Show the figure with subplots
plt.show()
```

```python
# Create a figure and a subplot
fig, ax = plt.subplots()

# Plot data with a label for the legend
ax.plot([0, 1, 2], [10, 20, 30], label='Line 1')
ax.plot([0, 1, 2], [30, 20, 10], label='Line 2')

# Set titles and labels
ax.set_title('Main Title')
ax.set_xlabel('X Axis Label')
ax.set_ylabel('Y Axis Label')

# Show the legend
ax.legend()

# Show the figure
plt.show()
```

```python
# Create a figure and a subplot
fig, ax = plt.subplots()

# Plot data
ax.plot([0, 1, 2], [10, 20, 30])

# Customize axes ticks
ax.set_xticks([0, 1, 2])
ax.set_yticks([10, 20, 30, 40])

# Add gridlines
ax.grid(True, which='both', axis='both', color='gray', linestyle='--', linewidth=0.5)

# Set axis limits
ax.set_xlim(0, 2)
ax.set_ylim(10, 40)

# Show the figure
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button

# Setup a plot
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.25, bottom=0.25)  # Make room for the slider

def f (par1, par2):
    return par1*par2


# Assume we have a function f that depends on a parameter 'a'
a0 = 5
x = np.arange(0, 1, 0.001)
y = f(a0, x)
line, = plt.plot(x, y, lw=2)

# Add slider for adjusting the parameter 'a'
axcolor = 'lightgoldenrodyellow'
ax_a = plt.axes([0.25, 0.1, 0.65, 0.03], facecolor=axcolor)

slider_a = Slider(ax_a, 'Param a', 0.1, 30.0, valinit=a0)

# Update the plot with the new value of 'a'
def update(val):
    line.set_ydata(f(slider_a.val, x))
    fig.canvas.draw_idle()

slider_a.on_changed(update)

# Add a button for resetting the parameters
resetax = plt.axes([0.8, 0.025, 0.1, 0.04])
button = Button(resetax, 'Reset', color=axcolor, hovercolor='0.975')

def reset(event):
    slider_a.reset()

button.on_clicked(reset)

plt.show()
```
